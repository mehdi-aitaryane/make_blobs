```python
from datasets import make_blobs, print_shape, count_occurances
from plots import scatter2D, scatter3D
```

```python
X, y = make_blobs(n_samples=1000, n_features=2, n_centers=3, balanced=True, random_state=111)
```

```python
print_shape(X, y)
```
